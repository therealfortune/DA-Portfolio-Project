```python
import pandas as pd
import numpy as np
```

```python
people = {
    'first' : ['Fortune', 'John', 'Jane'],
    'last' : ['Nwankwo', 'Doe', 'Doe'],
    'email' : ['FortuneNwankwo@email.com', 'JohnDoe@email.com', 'JaneDoe@email.com']
}
```

```python
df = pd.DataFrame(people)
```

```python
filt = (df['last'] == 'Doe') | (df['first'] == "John")
```

```python
df.loc[~filt, 'email']
```

```python
df.columns = ['First Name', 'Last Name', 'Email']
```
