```python
country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum()).head(2)
```

```python
country_reps = df['Country'].value_counts()
country_reps.head(2)
```

```python
country_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_python.head(2)
```

```python
 python_df = pd.concat([country_reps, country_python], axis='columns')
 python_df
```

```python
python_df.rename(columns={'Country':'NumOfRespondents', 'LanguagesWorkedWith': 'NumKnowsPython'}, inplace=True)
```

```python
python_df
```

```python
python_df['PctKnowsPython'] = (python_df['LanguageWorkedWith']/python_df['NumOfRespondents'])*100
python_df
```

```python
country_html = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('HTML').sum())
country_html
```

```python
html_df = pd.concat([country_reps , country_html], axis='columns')
html_df
```

```python
html_df.rename(columns={'Country':'TotalDevelopers', 'LanguageWorkedWith':'HtmlDevelopers'}, inplace=True)
```

```python
html_df
```

```python
html_df['HtmlPct'] = (html_df['HtmlDevelopers']/html_df['TotalDevelopers'])*100
```

```python
html_df.sort_values(by='HtmlPct', ascending=False)
```
